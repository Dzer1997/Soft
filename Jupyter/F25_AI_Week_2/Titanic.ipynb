```python
bins = [0, 10, 50, 100, 200, 1000]
labels = ['0-10', '10-50', '50-100', '100-200', '200+']

data['FareGroup'] = pd.cut(data['fare'], bins=bins, labels=labels)

print(data[['fare', 'FareGroup']].head())
```

```python
survival_by_fare = data.groupby('FareGroup')['survived'].mean()
print(survival_by_fare)
```

```python
sns.barplot(x=survival_by_fare.index, y=survival_by_fare.values)
plt.title('Survival Rate by Ticket Price Group')
plt.xlabel('Ticket Price Group')
plt.ylabel('Survival Rate')
plt.show()

# The chart is a barplot that shows the survival rate depending 
# on how much you spend
```
