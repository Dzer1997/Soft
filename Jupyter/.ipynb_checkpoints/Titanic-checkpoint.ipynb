```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("titanic")
data.head()
print(data['fare'].isnull().sum())
avg_fare_by_class = data.groupby('pclass')['fare'].mean()
print(avg_fare_by_class)
```

```python
sns.barplot(x=avg_fare_by_class.index, y=avg_fare_by_class.values)
plt.title('Average ticket price by class')
plt.xlabel('class')
plt.ylabel('Average ticket price ($)')
plt.show()
```

```python
bins = ['0', '10', '50', '100', '200', '1000']
labels = ['0-10', '10-50', '50-100', '100-200', '200+']

data['FareGroup'] = pd.cut(data['fare'], bins.bins, labels,labels, right=False)

print(data[['fare', 'FareGroup']].head())
```
