```python
import pandas as pd
df = pd.read_csv("output.csv")
print(df)

# Binning goals into categories
def categorize_goals(goals):
    if goals <= 1:
        return "Low"
    elif goals <= 3:
        return "Medium"
    else:
        return "High"

df["HomeGoalsCategory"] = df["HomeGoals"].apply(categorize_goals)
df["AwayGoalsCategory"] = df["AwayGoals"].apply(categorize_goals)

# One-hot encoding for the 'Result' column
df = pd.get_dummies(df, columns=["Result"], prefix="Result")

# Display the modified DataFrame
print(df)
```

```python
home_mean = df['HomeGoals'].mean()
home_std = df['HomeGoals'].std()

away_mean = df['AwayGoals'].mean()
away_std = df['AwayGoals'].std()

# Display the mean and std
print(home_mean)
print(home_std)
print(away_mean)
print(away_mean)
```

```python
def count_category_frequencies(column_name):
    return df[column_name].value_counts()

category_frequencies = {col: count_category_frequencies(col) for col in ["HomeGoalsCategory", "AwayGoalsCategory"]}

# Display the modified DataFrame and category frequencies
print(df)
print("\nCategory Frequencies:")
for category, freq in category_frequencies.items():
    print(f"{category}:\n{freq}\n")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))

# Box plots for numerical features
plt.subplot(1, 2, 1)
sns.boxplot(data=df[["HomeGoals", "AwayGoals"]])
plt.title("Box Plot of Goals")

```

```python

# Bar plots for categorical features
plt.subplot(1, 2, 2)
for category, freq in category_frequencies.items():
    sns.barplot(x=freq.index, y=freq.values)
    plt.title(f"Distribution of {category}")
    plt.xlabel(category)
    plt.ylabel("Count")
    plt.show()
```

```python
# Selecting relevant columns for prediction
selected_columns = ["HomeGoals", "AwayGoals", "HomeGoalsCategory", "AwayGoalsCategory", "Result_H"]
df_selected = df[selected_columns]

# Aggregating data using groupby
agg_df = df_selected.groupby(["HomeGoalsCategory", "AwayGoalsCategory"])[["HomeGoals", "AwayGoals"]].mean()

# Display results
print("Selected DataFrame:")
print(df_selected)
print("\nAggregated Mean Goals by Home and Away Goals Category:")
print(agg_df)
print("\nCategory Frequencies:")
```

```python
# Scatter plot to visualize relationships between numerical variables
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["HomeGoals"], y=df["AwayGoals"])
plt.title("Scatter Plot of HomeGoals vs. AwayGoals")
plt.xlabel("Home Goals")
plt.ylabel("Away Goals")
plt.show()

# Pairplot to visualize pairwise relationships
sns.pairplot(df, vars=["HomeGoals", "AwayGoals"])
plt.show()
```
